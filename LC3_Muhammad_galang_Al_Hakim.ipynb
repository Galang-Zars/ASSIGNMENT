**2.Apa saja jenis tindak kejahatan yang terjadi di kota Austin selama tahun 2016?**

```python
crime_type = client.query('''
SELECT DISTINCT(primary_type)
FROM `bigquery-public-data.austin_crime.crime`
WHERE year = 2016;
''').to_dataframe()

crime_type
```

**INSIGHT : Terdapat 7 crime type (rape, theft, robbery, aggravated assault, burglary, auto theft dan murder) yang terjadi di Kota Austin selama tahun 2016**

**3. Berapa banyak kasus tindak kejahatan berdasarkan kasus pembersihannya (clearance status) selama tahun 2016?**

```python
status_kriminal = client.query('''
SELECT EXTRACT(YEAR FROM clearance_date) AS year,
       clearance_status, COUNT(clearance_status) AS jumlah_status_kriminal
FROM `bigquery-public-data.austin_crime.crime`
WHERE EXTRACT(YEAR FROM clearance_date) = 2016
GROUP BY year, clearance_status
ORDER BY year
''').to_dataframe()

status_kriminal
```

**INSIGHT : Dari hasil analisa diatas terdapat Jumlah status kriminal sebanyak 5080  yang Cleared by Arrest, Jumlah status kriminal sebanyak 984 yang Cleared by Exception dan terdapat Jumlah status kriminal sebanyak 29673 yang Not Cleared**

**4. Jenis kejahatan apa saja yang paling banyak belum terselesaikan kasusnya di tahun 2016?**

```python
jenis_kejahatan_not_cleared = client.query('''
SELECT EXTRACT(YEAR FROM timestamp) AS year,
       primary_type, COUNT(primary_type) AS jumlah_kriminal
FROM `bigquery-public-data.austin_crime.crime`
WHERE EXTRACT(YEAR FROM timestamp) = 2016 AND clearance_status = "Not cleared"
GROUP BY year, primary_type
ORDER BY year, jumlah_kriminal DESC
''').to_dataframe()

jenis_kejahatan_not_cleared
```

**INSIGHT : kasus yang belum terselesaikan dan paling banyak didominasi dengan jenis kejahatan Theft sebanyak 21894 kasus diikuti dengan Burglary sebanyak 4440 kasus, Auto Theft sebanyak 1540, Aggravated Assault sebanyaka 919 kasus, Robbery sebanyak 560 kasus, Rape sebanyak 361 kasus dan murder sebanyak1 kasus yang belum terselesaikan.

**5. Di bulan apa di tahun 2016 kasus pencurian (theft) sangat banyak terjadi?**

```python
bulannya_theft = client.query('''
SELECT EXTRACT(YEAR FROM timestamp) AS year,
       EXTRACT(MONTH FROM timestamp) AS month,
       primary_type, COUNT(primary_type) AS jumlah_pencurian
FROM `bigquery-public-data.austin_crime.crime`
WHERE EXTRACT(YEAR FROM timestamp) = 2016 AND primary_type = 'Theft'
GROUP BY year, month, primary_type
ORDER BY year, jumlah_pencurian DESC
''').to_dataframe()

bulannya_theft
```

**INSIGHT : dari hasil analaisis data terdapat di Bulan April yang memiliki jumlah kriminal jenis Theft terbanyak, yaitu sebanyak 2282 kasus.**

**6.Di distrik apa yang paling banyak terjadi kasus pencurian selama 2016?**

```python
distrik_theft = client.query('''
SELECT EXTRACT(YEAR FROM timestamp) AS year,
       district, COUNT(district) AS paling_banyak_Kasus
FROM `bigquery-public-data.austin_crime.crime`
WHERE EXTRACT(YEAR FROM timestamp) = 2016
GROUP BY year, district
ORDER BY year, district DESC
''').to_dataframe()

distrik_theft
```

**INSIGHT : idstrik yang paling banyak terdapat kasus yaitu distrik F**
